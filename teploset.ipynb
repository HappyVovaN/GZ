```python
#Вывод графика
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))

Tnv = [i for i in range(-40,38)]
Tpr = [120 for i in range(-40,-28)]
for i in range(-28,8):
    Tpr.append(-0.8273*i+96.835)
for i in range(8,38):
    Tpr.append(90)
Tobr = [70 for i in range(-40,-28)]
for i in range(-28,8):
    Tobr.append(-0.9639*i+43.012)
for i in range(8,38):
    Tobr.append(0.1724*i+33.621)
plt.plot(Tnv,Tpr, color = 'red')
plt.plot(Tnv,Tobr, color = 'blue')
plt.xlabel('Tнв')
plt.ylabel('T')
plt.legend(['Прямая','Обратная'])
plt.show()
# t=36
# i=t+40
# print(Tpr[i],Tobr[i])


#Функция определения температуры в теплосети
def Tset(Tnv):
    if Tnv <=-28:
        Tpr=120
        Tobr=70
    if Tnv>-28 and Tnv<=8:
        Tpr=-0.8273*Tnv+96.835
        Tobr=-0.9639*Tnv+43.012
    if Tnv>8:
        Tpr=90
        Tobr=0.1724*Tnv+33.621
    return Tpr, Tobr
print(Tset(5))
```
